```python
import numpy as np
import pandas as pd
import math
from math import sqrt
```

```python
# 데이터불러오기
user_thema_data_two = pd.read_csv('../data/dataset2/user_thema_data_two.csv')
user_thema_data_two.head()
```

```python
# list형식 쪼개기
userid     = []
themacode  = []

for i in range(len(user_thema_data_two['themaCode'])) :
    for j in user_thema_data_two.iloc[i,1].split() :
        userid.append(user_thema_data_two.iloc[i,0])
        j = j.replace("'", "")
        j = j.replace(",", "")
        j = j.replace("[", "")
        j = j.replace("]", "")
        themacode.append(j)
data = pd.DataFrame([userid, themacode],
                   index = ['userId', 'themacode'])
data = data.T
```

```python
# pivot table로 변환하기 위해 임의로 rating=1 부여
data['rating']=1
data
```

```python
# pivot table변환
theme_data = data.pivot_table('rating', index='userId', columns='themacode').fillna(0).astype(int)
theme_data
```

```python
#테마X이름 행렬로 만들기위해 Transpose
theme_data=pd.DataFrame(data=theme_data.T)

#데이터 프레임 딕셔너리 구조로 변경
theme_data_dic=theme_data.to_dict()
theme_data_dic
```

```python
# dictionary에 nan값 함수 
def dropna(data): # name : i , movie : j 
    for i in data:
        for j in data[i]:
            name=i
            theme=j
            value=data[i][j]
        data[i]={travel: value for travel, value in data[i].items() if pd.isnull(value)==False}
    return data

# 딕셔너리 nan 값 제거 
theme_data_dic_drop=dropna(theme_data_dic)
theme_data_dic_drop
```

```python
def sim_pearson(data, n1, n2): 
    sumX=0
    sumY=0
    sumSqX=0 # x 제곱합 
    sumSqY=0 # y 제곱합 
    sumXY=0 # XY 합
    global cnt # 테마 갯수
    cnt=0
    for i in data[n1]:
        if i in data[n2]:
            sumX+=data[n1][i]
            sumY+=data[n2][i]
            sumSqX+=pow(data[n1][i],2)
            sumSqY+=pow(data[n2][i],2)
            sumXY+=(data[n1][i])*(data[n2][i])
            cnt+=1
            global num
            global den
            num=sumXY-((sumX*sumY)/cnt)
            den= (sumSqX-(pow(sumX,2)/cnt))*(sumSqY-(pow(sumY,2)/cnt))
    return num/sqrt(den+0.00001) # 분모=0방지
```

```python
def top_match(data, name, rank=3, simf=sim_pearson):
    simList=[] # 유사한 테마들이 저장될 리스트
    for i in data:
        if name!=i: # 자기 자신 제외
            simList.append((simf(data, name, i),i))
    simList.sort() # 오름차순
    simList.reverse() # 내림차순
    return simList[:rank]
    
    # 내가 선택한 테마를 바탕으로 나와의 성향이 비슷하게 나온 사람 3명
print(top_match(theme_data_dic_drop,"aabkmf32@google.com",3))
```
